```python
# ----------------------------------------
# Exercise 1: Introduction to Data Analysis
# ----------------------------------------
# Write a short essay about data analysis (not executable in code)
essay = """
Data analysis is the process of examining, cleaning, transforming, and modeling data
to discover useful information, draw conclusions, and support decision-making.
It is important because it helps organizations make informed choices, identify trends,
and improve operational efficiency. Some applications include business analytics,
healthcare analytics, and finance risk assessment.
"""
print("Exercise 1: Introduction Essay\n", essay)

# ----------------------------------------
# Exercise 2: Dataset Loading and Initial Analysis
# ----------------------------------------
import pandas as pd

# Example dataset from Kaggle (replace URL with actual path if needed)
# For demonstration, using a sample CSV from GitHub (replace with Kaggle dataset locally in Colab)
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

# Display first few rows
print("\nExercise 2: First 5 rows of dataset")
print(df.head())

# Brief description of dataset
print("\nDataset description:")
print(df.info())

# ----------------------------------------
# Exercise 3: Identifying Data Types
# ----------------------------------------
# Classify each column as quantitative or qualitative
data_types = {}
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        data_types[col] = 'Quantitative'
    else:
        data_types[col] = 'Qualitative'

print("\nExercise 3: Data Types")
for col, dtype in data_types.items():
    print(f"{col}: {dtype}")

# ----------------------------------------
# Exercise 4: Exploring Data Types
# ----------------------------------------
# Load Iris dataset already done in Exercise 2
# Identify qualitative and quantitative columns
quantitative_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
qualitative_cols = df.select_dtypes(exclude=['int64', 'float64']).columns.tolist()

print("\nExercise 4: Exploring Data Types")
print("Quantitative columns:", quantitative_cols)
print("Qualitative columns:", qualitative_cols)

# ----------------------------------------
# Exercise 5: Basic Data Analysis With Google Colab
# ----------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate mean, median, mode of a quantitative column (example: sepal_length)
mean_val = df['sepal_length'].mean()
median_val = df['sepal_length'].median()
mode_val = df['sepal_length'].mode()[0]

print("\nExercise 5: Basic Analysis")
print(f"Mean sepal_length: {mean_val}")
print(f"Median sepal_length: {median_val}")
print(f"Mode sepal_length: {mode_val}")

# Simple histogram using matplotlib
plt.hist(df['sepal_length'], bins=10, color='skyblue', edgecolor='black')
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()

# Using seaborn for a bar plot by species
sns.barplot(x='species', y='sepal_length', data=df)
plt.title("Average Sepal Length by Species")
plt.show()

# ----------------------------------------
# Exercise 6: Basic Observation Skills in Data Analysis
# ----------------------------------------
# Example of basic observation: compare sepal_length and sepal_width
print("\nExercise 6: Basic Observation")
correlation = df['sepal_length'].corr(df['sepal_width'])
print(f"Correlation between sepal_length and sepal_width: {correlation}")

# Identify columns interesting for analysis
interesting_cols = ['sepal_length', 'petal_length', 'species']
print("Columns chosen for observation:", interesting_cols)
```
