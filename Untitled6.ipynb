```python
# Modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données
sales_df = pd.read_csv('/content/sales_data.csv')

# --- Exercise 1 : Understanding Data Visualization ---
print("Exercise 1 :")
print("La visualisation des données est essentielle pour interpréter rapidement les tendances, comparer des catégories, détecter des anomalies et communiquer les insights de manière claire.")
print("Un graphe linéaire (line plot) permet de visualiser l'évolution d'une variable continue dans le temps ou selon une séquence.")

# --- Exercise 2 : Line Plot for Temperature Variation ---
# Exemple fictif de températures
temps = [72, 74, 76, 80, 82, 78, 75]
jours = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

plt.figure(figsize=(8,5))
plt.plot(jours, temps, marker='o', color='b')
plt.title('Temperature Variation Over a Week')
plt.xlabel('Day')
plt.ylabel('Temperature (°F)')
plt.grid(True)
plt.show()

# --- Exercise 3 : Monthly Sales Bar Chart ---
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [5000, 5500, 6200, 7000, 7500]

plt.figure(figsize=(8,5))
plt.bar(months, sales, color='orange')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.show()

# --- Exercise 4 : Data Visualisation from CSV ---
# Calculer la quantité totale vendue
total_quantity = sales_df['Quantity'].sum()
print("Total quantity sold:", total_quantity)

# Catégorie avec le revenu le plus élevé
sales_df['Revenue'] = sales_df['Quantity'] * sales_df['Price']
top_category = sales_df.groupby('Category')['Revenue'].sum().idxmax()
top_category_revenue = sales_df.groupby('Category')['Revenue'].sum().max()
print("Category with highest revenue:", top_category)
print("Revenue for this category:", top_category_revenue)

# Revenu moyen par vente
avg_revenue = sales_df['Revenue'].mean()
print("Average revenue per sale:", avg_revenue)

# Bar chart trimestriel
# Assumer une colonne 'Quarter' existe dans ton dataset, sinon créer fictif:
if 'Quarter' not in sales_df.columns:
    sales_df['Quarter'] = [1,1,2,2,3,3,4,4][:len(sales_df)]

quarterly_revenue = sales_df.groupby('Quarter')['Revenue'].sum()
quarterly_revenue.plot(kind='bar', figsize=(8,5), color='green')
plt.title('Quarterly Revenue')
plt.xlabel('Quarter')
plt.ylabel('Revenue')
plt.show()

# --- Exercise 5 : Matplotlib Plots ---
# Line plot y = x^2
import numpy as np
x = np.arange(-10,11)
y = x**2

plt.figure(figsize=(8,5))
plt.plot(x, y)
plt.title("Plot of y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()

# Bar chart 4 produits A, B, C, D
products = ['A', 'B', 'C', 'D']
sales_values = [15, 30, 45, 20]

plt.figure(figsize=(6,4))
plt.bar(products, sales_values, color=['red','blue','green','purple'])
plt.title('Sales per Product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()

# Pie chart pour les fruits préférés
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
percentages = [40, 30, 20, 10]

plt.figure(figsize=(6,6))
plt.pie(percentages, labels=fruits, autopct='%1.0f%%', colors=['red','yellow','pink','brown'])
plt.title('Favorite Fruits Distribution')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données
sales_df = pd.read_csv('/content/sales_data.csv')

# 1️⃣ Exercice 1 : Understanding Data Visualization
# Réponse : La visualisation permet de comprendre rapidement les tendances, anomalies, et relations dans les données.
# Une ligne (line plot) est utile pour montrer les variations dans le temps ou les séquences.

# 2️⃣ Exercice 2 : Line Plot for Temperature Variation
temps = [72, 74, 76, 80, 82, 78, 75]
jours = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
plt.plot(jours, temps, marker='o')
plt.title('Temperature Variation Over a Week')
plt.xlabel('Day')
plt.ylabel('Temperature (°F)')
plt.show()

# 3️⃣ Exercice 3 : Monthly Sales Bar Chart
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [5000, 5500, 6200, 7000, 7500]
plt.bar(months, sales, color='orange')
plt.title('Monthly Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.show()

# 4️⃣ Exercice 4 : Data Visualization from CSV
# Total quantity sold
total_quantity = sales_df['quantity'].sum()
print("Total quantity sold:", total_quantity)

# Category with highest revenue
highest_rev_category = sales_df.groupby('category')['revenue'].sum().idxmax()
highest_rev_amount = sales_df.groupby('category')['revenue'].sum().max()
print(f"Category with highest revenue: {highest_rev_category}, Revenue: {highest_rev_amount}")

# Average revenue per sale
avg_revenue = sales_df['revenue'].mean()
print("Average revenue per sale:", avg_revenue)

# Bar chart by quarter
sales_df['date'] = pd.to_datetime(sales_df['date'])
sales_df['quarter'] = sales_df['date'].dt.quarter
quarter_revenue = sales_df.groupby('quarter')['revenue'].sum()
quarter_revenue.plot(kind='bar', color='skyblue')
plt.title('Total Revenue per Quarter')
plt.xlabel('Quarter')
plt.ylabel('Revenue ($)')
plt.show()

# 5️⃣ Exercice 5 : Matplotlib line, bar, pie charts
# Line plot
x = list(range(-10, 11))
y = [i**2 for i in x]
plt.plot(x, y)
plt.title('y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Bar chart for products
products = ['A', 'B', 'C', 'D']
values = [15, 30, 45, 20]
plt.bar(products, values, color=['red','green','blue','purple'])
plt.title('Weekly Sales of Products')
plt.xlabel('Products')
plt.ylabel('Units Sold')
plt.show()

# Pie chart for fruits
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [40, 30, 20, 10]
plt.pie(sizes, labels=fruits, autopct='%1.1f%%', colors=['red','yellow','pink','brown'])
plt.title('Favorite Fruits Distribution')
plt.show()
```
