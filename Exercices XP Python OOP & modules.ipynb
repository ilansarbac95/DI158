```python
# ----------------------------------------
# Exercise 1: Currencies
# ----------------------------------------
class Currency:
    def __init__(self, currency, amount):
        self.currency = currency
        self.amount = amount

    def __int__(self):
        return self.amount

    def __repr__(self):
        return f"{self.amount} {self.currency}"

    def __add__(self, other):
        if isinstance(other, Currency):
            if self.currency == other.currency:
                return Currency(self.currency, self.amount + other.amount)
            else:
                raise TypeError(f"Cannot add between Currency type {self.currency} and {other.currency}")
        elif isinstance(other, int):
            return Currency(self.currency, self.amount + other)
        else:
            raise TypeError("Unsupported type for addition")

    def __iadd__(self, other):
        if isinstance(other, Currency) and self.currency == other.currency:
            self.amount += other.amount
        elif isinstance(other, int):
            self.amount += other
        else:
            raise TypeError("Unsupported type for addition")
        return self

# Test Currency
c1 = Currency('dollar', 5)
c2 = Currency('dollar', 10)
c3 = Currency('shekel', 1)
c4 = Currency('shekel', 10)

print(c1 + 5)
print(c1 + c2)
print(c1)
c1 += 5
print(c1)
c1 += c2
print(c1)
# print(c1 + c3)  # This would raise TypeError

# ----------------------------------------
# Exercise 2: Import
# ----------------------------------------
# Step 1: Create func.py
# def sum_two(a, b):
#     return a + b
# Step 2: Create exercise_one.py
# from func import sum_two
# print(sum_two(3, 4))

# For Colab, we can simulate the import inline
def sum_two(a, b):
    return a + b

print(sum_two(3, 4))

# ----------------------------------------
# Exercise 3: String Module
# ----------------------------------------
import string
import random

all_letters = string.ascii_letters  # Uppercase + lowercase
random_str = ''.join(random.choice(all_letters) for _ in range(5))
print("Random string:", random_str)

# ----------------------------------------
# Exercise 4: Current Date
# ----------------------------------------
from datetime import datetime
now = datetime.now()
print("Current date and time:", now)

# ----------------------------------------
# Exercise 5: Amount of Time Left Until January 1st
# ----------------------------------------
from datetime import timedelta

def time_until_january():
    today = datetime.now()
    next_year = datetime(today.year + 1, 1, 1)
    diff = next_year - today
    print("Time until next January 1st:", diff)

time_until_january()

# ----------------------------------------
# Exercise 6: Birthday and Minutes
# ----------------------------------------
def birthday_minutes(birthdate):
    birthday = datetime.strptime(birthdate, "%Y-%m-%d")
    now = datetime.now()
    delta = now - birthday
    minutes = delta.total_seconds() / 60
    print(f"Minutes since {birthdate}: {minutes}")

birthday_minutes("2000-01-01")

# ----------------------------------------
# Exercise 7: Faker Module
# ----------------------------------------
!pip install faker -q
from faker import Faker
fake = Faker()

def generate_users(n):
    users = []
    for _ in range(n):
        users.append({
            "name": fake.name(),
            "email": fake.email()
        })
    return users

user_list = generate_users(5)
print(user_list)
```
