```python
import json

# Given JSON string from the exercise
sample_json = """
{
  "company": {
    "employee": {
      "name": "emma",
      "payable": {
        "salary": 7000,
        "bonus": 800
      }
    }
  }
}
"""


def main():
    # Step 1: Load the JSON string into a Python dictionary
    data = json.loads(sample_json)

    # Step 2: Access the nested "salary" key
    salary = data["company"]["employee"]["payable"]["salary"]
    print(f"Current salary is: {salary}")

    # Step 3: Add the "birth_date" key to the employee dictionary
    # Replace this date with any real date you want
    data["company"]["employee"]["birth_date"] = "1995-08-15"

    # Step 4: Save the modified JSON to a file
    output_file = "employee_data.json"

    with open(output_file, "w", encoding="utf-8") as file:
        json.dump(data, file, indent=4)

    print(f"Modified JSON has been saved to {output_file}")


if __name__ == "__main__":
    main()
```
