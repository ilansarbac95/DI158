```python
# Import des biblioth√®ques
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# √âtape 1 : G√©n√©ration des donn√©es (Temp√©ratures al√©atoires entre -5 et 35¬∞C)
np.random.seed(42)  # pour reproductibilit√©
temps = np.random.uniform(low=-5, high=35, size=(10, 12))  # 10 villes x 12 mois

# Cr√©ation du DataFrame
cities = [f'City {i+1}' for i in range(10)]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

df = pd.DataFrame(data=temps, index=cities, columns=months)

# √âtape 2 : Analyse des donn√©es

# Moyenne annuelle par ville
df['Annual Avg Temp'] = df.mean(axis=1)

# Ville avec temp√©rature moyenne la plus √©lev√©e et la plus basse
hottest_city = df['Annual Avg Temp'].idxmax()
coldest_city = df['Annual Avg Temp'].idxmin()

# Affichage des moyennes
print("üå°Ô∏è Moyennes annuelles par ville:")
print(df['Annual Avg Temp'].round(2))

print(f"\nüî• Ville la plus chaude : {hottest_city} ({df.loc[hottest_city, 'Annual Avg Temp']:.2f} ¬∞C)")
print(f"‚ùÑÔ∏è Ville la plus froide : {coldest_city} ({df.loc[coldest_city, 'Annual Avg Temp']:.2f} ¬∞C)")

# √âtape 3 : Visualisation
plt.figure(figsize=(12, 6))
for city in cities:
    plt.plot(months, df.loc[city, months], label=city)

plt.title("üìà Monthly Temperature Trends for Each City")
plt.xlabel("Month")
plt.ylabel("Temperature (¬∞C)")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Summary of the Analysis

- This simulation generates monthly temperature data for 10 cities, with random values ranging from -5¬∞C to 35¬∞C.
- The resulting DataFrame includes monthly temperatures and the calculated annual average temperature for each city.
- The hottest city is **City X**, while the coldest is **City Y**.
- The visualization clearly highlights seasonal trends and differences in climate across the cities.
