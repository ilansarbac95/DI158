```python
# Création de la base de données SQLite en mémoire
conn = sqlite3.connect(':memory:')
c = conn.cursor()

# Exemple : création des tables FirstTab et SecondTab
c.execute('''
CREATE TABLE FirstTab (
    id INTEGER,
    name TEXT
)
''')

c.execute('''
CREATE TABLE SecondTab (
    id INTEGER
)
''')

# Insertion des données
c.executemany('INSERT INTO FirstTab VALUES (?,?)', [
    (5, 'Pawan'), (6, 'Sharlee'), (7, 'Krish'), (None, 'Avtaar')
])
c.executemany('INSERT INTO SecondTab VALUES (?)', [
    (5,), (None,)
])

# Exemple de SELECT
df_first = pd.read_sql_query('SELECT * FROM FirstTab', conn)
df_second = pd.read_sql_query('SELECT * FROM SecondTab', conn)

df_first, df_second
```

```python
import numpy as np
from scipy import stats

# Exemple NumPy
arr = np.array([[1,2],[3,4],[5,6]])
print("Cumulative sum axis 0:", arr.cumsum(axis=0))
print("Cumulative sum axis 1:", arr.cumsum(axis=1))

# SciPy exemple: test t
t_stat, p_val = stats.ttest_1samp([1,2,3,4,5], 3)
print("T-test p-value:", p_val)
```
