```python
import random


WORDS_FILE = "words.txt"


def get_words_from_file(file_path):
    """
    Read all the words from a text file and return them as a list.

    Each word is expected to be separated by whitespace (spaces, newlines, etc.).
    """
    with open(file_path, "r", encoding="utf-8") as file:
        content = file.read()

    # Split the whole content into a list of words
    words = content.split()
    return words


def get_random_sentence(length):
    """
    Generate a random sentence of a given length (number of words).

    The words are picked randomly from the word list in WORDS_FILE.
    The final sentence is returned in lowercase.
    """
    words = get_words_from_file(WORDS_FILE)

    # Randomly pick `length` different words from the list
    chosen_words = random.sample(words, length)

    # Join the words into a sentence and convert to lowercase
    sentence = " ".join(chosen_words).lower()

    # Add a period at the end if there is none
    if not sentence.endswith("."):
        sentence += "."

    return sentence


def main():
    """
    Main function that asks the user for a sentence length,
    validates the input and then prints a random sentence.
    """
    print("Random Sentence Generator")
    print("-------------------------")
    print("This program generates a random sentence from a word list.")

    user_input = input("How many words should the sentence contain (between 2 and 20)? ")

    # Validate that the user typed an integer
    try:
        length = int(user_input)
    except ValueError:
        print("Error: please enter a whole number.")
        return

    # Check the allowed range
    if length < 2 or length > 20:
        print("Error: the number must be between 2 and 20.")
        return

    # Everything is fine: generate and print the sentence
    sentence = get_random_sentence(length)
    print("\nHere is your random sentence:")
    print(sentence)


if __name__ == "__main__":
    main()
```
