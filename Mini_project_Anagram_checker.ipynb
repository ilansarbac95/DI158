```python
class AnagramChecker:
    def __init__(self, wordlist_path='sowpods.txt'):
        with open(wordlist_path, 'r') as f:
            self.words = [line.strip().upper() for line in f.readlines()]

    def is_valid_word(self, word):
        return word.upper() in self.words

    def get_anagrams(self, word):
        word = word.upper()
        return [w for w in self.words if w != word and sorted(w) == sorted(word)]

    @staticmethod
    def is_anagram(word1, word2):
        return sorted(word1.upper()) == sorted(word2.upper()) and word1.upper() != word2.upper()
```

```python
from anagram_checker import AnagramChecker

def clean_input(text):
    return text.strip()

def is_valid_input(text):
    return text.isalpha() and ' ' not in text

def display_result(word, anagrams):
    print("\nYOUR WORD : \"" + word.upper() + "\"")
    print("This is a valid English word.")
    if anagrams:
        print("Anagrams for your word:", ', '.join(anagrams))
    else:
        print("No anagrams found.")

def main():
    checker = AnagramChecker()
    while True:
        print("\nüü¶ MENU:")
        print("1. Enter a word")
        print("2. Exit")
        choice = input("Choose an option (1 or 2): ").strip()

        if choice == '2':
            print("Bye üëã")
            break

        elif choice == '1':
            word = clean_input(input("Enter a single word: "))

            if not is_valid_input(word):
                print("‚ùå Invalid input. Please enter a single alphabetical word without spaces or numbers.")
                continue

            if not checker.is_valid_word(word):
                print("‚ùå Word not found in English dictionary.")
                continue

            anagrams = checker.get_anagrams(word)
            display_result(word, anagrams)

        else:
            print("‚ùå Invalid choice. Please enter 1 or 2.")

if __name__ == '__main__':
    main()
```
